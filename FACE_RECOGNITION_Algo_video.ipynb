```python
import face_recognition
import cv2
import os
import time

#Dossier

KNOWN_FACES_DIR = "./Desktop/face_algo _Video/known_faces"
#UNKNOWN_FACES_DIR = "./Desktop/face_algo _Video/unknown_faces"
tolerance = 0.6 #valeur: entre 0 et 1. //permet de modifier la sensibilité de l'image
frame_epaisseur = 3
font_epaisseur =2
MODEL = "Cnn"# Convolutional Neural Network

video =cv2.VideoCapture(0)
#Chargement des visage connue
print("Loading known faces avec les noms associés....")

#listes
known_faces = []
known_name = []

for name in os.listdir(KNOWN_FACES_DIR):
	#chargement de tous les fichiers d'images de visage connue contenant dans le dossier
	for filename in os.listdir(f'{KNOWN_FACES_DIR}/{name}'):
		#une fois dans le dossier on charge les images
		image = face_recognition.load_image_file(f"{KNOWN_FACES_DIR}/{name}/{filename}")
		#on va encoder l'image: # Get 128-dimension face encoding
		encoding = face_recognition.face_encodings(image)[0]#selection du 1er image seulement
		#ajout au listes declarés
		known_faces.append(encoding)
		known_name.append(name)

#À ce stade, nous sommes prêts à vérifier les visages inconnus, puis à essayer d'identifier ces visages!
while True:
	ret,image = video.read()
	location = face_recognition.face_locations(image, model =MODEL)# on va essayer de recuperer tous les carré sur l'image inconnue(pour eviter tout confusion)
	#encodings
	encoding = face_recognition.face_encodings(image,location)
	# on va maintenant dessiner sur l'image grace à cv2: rectangle,couleur
	#image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
	#on va faire une iterration sur : encoding et location des images que nous allons trouver afin de faire une COMPARAISON
	for face_encoding, face_loaction in zip(encoding,location): #2 variables: face_encoding pour comparson et face_loaction pour localiser l'image: en se basant sur son index
		results = face_recognition.compare_faces(known_faces,face_encoding, tolerance)#comparaison d'image connue avec inconnue encodée
		match =None
		if True in results: #s'il trouve une vraie(true) pour une image , il va nous fournir son index
			match = known_name[results.index(True)]
			print(f'match found : {match}')

			#maintenant nous allons dessiner le rectangle avec cv2: pour cela on a besoin de 2 coordonnées:top_left, bottom_right
			top_left = (face_loaction[3],face_loaction[0])
			bottom_right = (face_loaction[1],face_loaction[2])
			color = [0,255,0]
			#dessin du rectangle
			cv2.rectangle(image , top_left,bottom_right, color,frame_epaisseur)
			#dessin d'un petit rectangle
			top_left = (face_loaction[3],face_loaction[2])
			bottom_right = (face_loaction[1],face_loaction[2]+22)
			cv2.rectangle(image , top_left,bottom_right, color,cv2.FILLED)
			#insertion de text dans le rectangle
			cv2.putText(image, match, (face_loaction[3] + 10, face_loaction[2] + 15), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (12, 12, 12), font_epaisseur)

	cv2.imshow(filename,image)
	if cv2.waitKey(1) & 0xFF == ord("q"):
		break
```
